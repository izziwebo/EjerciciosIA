### Listas por Compresión

Sirve para crear nuevas listas a partir de iterables existentes en una sola línea de código. Se construye a partir de la siguiente estructura:

`nueva_lista = [expresión for elemento in iterable if condición]`

El siguiente ejemplo muestra una lista que considera números del 0 a 9, donde se considerarán en la lista solo aquellos que sean pares:

```python
ListaCompresion = [i for i in range(10) if i%2 == 0]
ListaCompresion
```

### Diccionarios
Son estructuras de datos que almacenan pares `clave:valor` encerrados entre llaves {}.

```python
Diccionario = {"Mark": {"Edad":19, "Ciudad": "Detroit", "Super": "Invencible"}, "Eve": {"Edad":18, "Ciudad": "Chicago", "Super": "Atom Eve"}}

print(Diccionario["Mark"]["Super"])
print(Diccionario["Eve"]["Super"])
```

### Condicionales

```python
for Clave, Valor in Diccionario.items():
    if Valor["Edad"] >= 18:
        print(f"{Valor['Super']} es mayor de edad.")
    elif Valor["Edad"] < 18:
        print(f"{Valor['Super']} no es mayor de edad.")
```

### Excepciones: Controlar muchas variables

El bloque `try` contiene el código que puede generar un error, mientras que `except` define las acciones a tomar si dicho error ocurre, evitando que el programa se detenga abruptamente.

```python
try:
    print(Diccionario["Mark"]["Altura"]) # Esto va a dar Error, porque no existe Altura en el Diccionario de Mark
except Exception as E: # Capturar el Error en "E"
    print("ERROR:", E) # Imprimir el Error
```

### Bucles

Existen dos tipos de bucles, preferiblemente utilizar `for`, aunque también existe `while`.

```python
for a in Diccionario.keys(): # Iterar sobre las Claves del Diccionario
    print("El civil", a, "es el superhéroe", Diccionario[a]["Super"])
```

### Lista por Compresión de Diccionario

En este caso seguimos la misma estructura, pero para cambiar las condiciones, dejaremos que sea superior a 18 la edad, así, solamente `Mark` podría entrar a esta lista según la condición.

```python
listCompresion = [i for i in Diccionario.keys()
                  if Diccionario[i]["Edad"] > 18]
listCompresion
```

### Arrays:

Son arreglos inmutables, no se pueden modificar sus elementos. Esa sería la diferencia con las listas que sí lo son.

```python
arreglo = np.array([1,2,3.5,4,5])
arreglo.dtype # Tipo de dato en el arreglo
# Según lo que exista, se ve que predomina, en este caso hay un puro float pero igual predomina por el resto.
```

Para ciertas operaciones como concatenación de arreglos, se utiliza `concat`, uniendo dos arreglos.

<b>Nota:</b> Cabe aclarar que debe existir algún predominante en caso de que se junten typos de datos. Por ejemplo si se combinan enteros con strings, entonces los strings van a predominar.

```python
array = np.array([1,2,3,4,5])
array2 = np.array(["Grande", "Erivo", "Bailey", "Goldblum", "Slater"])

array_combinado = np.concat([array,array2])
array_combinado
```

### Números Aleatorios
- Utilizando la función `seed`, con el código o número que se le representa, es una semilla, o sea una estructura ya construida y guardada. Lo que en sí, garantiza la reproducibilidad de este arreglo.

- Utilizando la función `rand`, esta considera una estructura `(filas, columnas)`, por lo que en este caso este arreglo tiene <b>1 fila</b> y <b>10 columnas</b>. Se consideran números aleatorios entre 0 y 1.

```python
np.random.seed(42)
random_array = np.random.rand(1,10)
random_array
```

Existen funciones para la <b>redimensión de los arreglos</b>, esto se logra con `reshape` siguiendo la misma estructura de `(filas, columnas)`.

```python
changed_array = random_array.reshape(2,5)
changed_array
```

```python
# Traspuesta del Arreglo
changed_array.T
```

### Unos y Ceros

- La funcion `ones` sirve para realizar un arreglo de solamente unos, pero siguiendo una estructura `(filas, columnas)`.

- La funcion `zeros` sirve para realizar un arreglo de solamente ceros, pero siguiendo una estructura `(filas, columnas)`.

```python
Unos = np.ones((2,5))
Unos
```

```python
Ceros = np.zeros((2,5))
Ceros
```

```python
# Operaciones o Ruido a los Arrays
print(Unos * 2 + changed_array)
```

```python
# Desviación Estándar
print(f"La desviación estándar es: {changed_array.std()}")

# Varianza
print(f"La varianza es: {changed_array.std() ** 2}")
```

### Creacion de Espacios Lineales

Primero tenemos `linspace`, que sirve para generar un vector o tensor de números equiespaciados dentro de un intervalo específico. Sigue la estructura a continuación:

`np.linspace(inicio, fin, cantidad)`

```python
np.linspace(0,10,5)
```

Otro es `arange`, el cual sirve para lo mismo, pero en vez de poner la cantidad se pone el valor del paso. Sigue la estructura a continuación:

`np.arange(inicio, fin, paso)`

```python
np.arange(0,10,0.5)
```
